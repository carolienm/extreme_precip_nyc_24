```python
#Reviewer plot
clrlist=['black','orange','blue']
fig, ax = plt.subplots(figsize=(8,6))
yr=[10,25,50]
for ll in range(8):
    if ll == 0:
        plt.plot(yr,dd[ll][1:4],linestyle='solid',color='black',linewidth=3)
    if ll == 1:
        plt.plot(yr,dd[ll][1:4],linestyle='solid',color='grey',linewidth=3)
    if ll == 2:
        plt.plot(yr,dd[ll][1:4],linestyle='dashed',color=clrlist[0],linewidth=3)
    if ll == 3:
        plt.plot(yr,dd[ll][1:4],linestyle='solid',color=clrlist[0],linewidth=3)
    if ll ==4:
        plt.plot(yr,dd[ll][1:4],linestyle='dashed',color=clrlist[1],linewidth=3)
    if ll == 5:
        plt.plot(yr,dd[ll][1:4],linestyle='solid',color=clrlist[1],linewidth=3)
    if ll ==6:
        plt.plot(yr,dd[ll][1:4],linestyle='dashed',color=clrlist[2],linewidth=3)
    if ll == 7:
        plt.plot(yr,dd[ll][1:4],linestyle='solid',color=clrlist[2],linewidth=3)
    #plt.scatter(yr,dd[ll][1:4])
plt.legend(data['model'])
plt.title('Risk- Average Temperature',weight='bold')
plt.ylabel('Expected Probability of Ida-like Event',weight='bold')
plt.xlabel('Planning Phase (Years from 2024)',weight='bold')
plt.xticks(yr)
plt.xlim([0,51])
plt.ylim([0,1.2])
plt.setp(ax.get_xticklabels(), fontsize=10,weight='bold')
plt.setp(ax.get_yticklabels(), fontsize=10,weight='bold')
plt.plot([10,10],[0,1.3],'--k')
plt.plot([25,25],[0,1.3],'--k')
plt.plot([50,50],[0,1.3],'--k')
```
